```python
import pandas as pd

# Load all parts
X_train = pd.read_csv('../../data/processed/X_train_landlord_scaled.csv')
X_test = pd.read_csv('../../data/processed/X_test_landlord_scaled.csv')
y_train = pd.read_csv('../../data/processed/y_train_landlord.csv')
y_test = pd.read_csv('../../data/processed/y_test_landlord.csv')

# Load original data with id column
original_data = pd.read_csv('../../data/processed/listings_with_categorical_encoding.csv')

# Combine X and y for train and test
train_combined = pd.concat([X_train, y_train], axis=1)
test_combined = pd.concat([X_test, y_test], axis=1)

# Combine all into one dataset
final_data = pd.concat([train_combined, test_combined], axis=0, ignore_index=True)

# Add id column from original data
final_data['id'] = original_data['id'].values

# Move id to first column
cols = ['id'] + [col for col in final_data.columns if col != 'id']
final_data = final_data[cols]

print(f"Final dataset shape: {final_data.shape}")

# Save
final_data.to_csv('../../data/finalized/numeric_final_data.csv', index=False)
print("Saved as numeric_final_data.csv")
```
